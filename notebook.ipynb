```python

import pandas as pd
import matplotlib.pyplot as plt
combined_df = pd.read_csv("birlestirilmis_veriler_clean.csv")
print(combined_df.head())
```

# Yeni Bölüm

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot steps on the first y-axis
ax1.plot(combined_df["date"], combined_df["steps"], label="Adım Sayısı", marker='o', color='blue')
ax1.set_xlabel("Tarih")
ax1.set_ylabel("Adım Sayısı", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a twin y-axis for temperature
ax2 = ax1.twinx()
ax2.plot(combined_df["date"], combined_df["avg_temp"], label="Ortalama Sıcaklık (°C)", marker='x', color='red')
ax2.set_ylabel("Ortalama Sıcaklık (°C)", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Rotate x-axis labels
ax1.tick_params(axis='x', rotation=45)

# Add title and layout adjustments
fig.suptitle("Günlük Adım Sayısı ve Ortalama Sıcaklık")
fig.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(combined_df["avg_temp"], combined_df["steps"], alpha=0.7, color='purple')
plt.xlabel("Ortalama Sıcaklık (°C)")
plt.ylabel("Adım Sayısı")
plt.title("Sıcaklık ve Adım Sayısı Dağılımı")
plt.grid(True)
plt.show()
```
